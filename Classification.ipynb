# Now we will go for the classification techniques

```
from sklearn.model_selection import train_test_split
x=df.drop(['price_range'],axis=1)
y=df['price_range']
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.3,random_state=0)
```

```
print(len(xtr),len(xts))
```

```
from sklearn.naive_bayes import GaussianNB

gnb=GaussianNB()

gnb.fit(xtr,ytr)

pred=gnb.predict(xts)
pred
```

```
from sklearn.metrics import accuracy_score
accuracy_score(yts,pred)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(yts,pred)
```

```
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=10000)
model.fit(xtr,ytr)
pred=model.predict(xts)
from sklearn.metrics import classification_report
print(classification_report(yts,pred))
```

```
confusion_matrix(yts,pred)
```

```
accuracy_score(yts,pred)
```
